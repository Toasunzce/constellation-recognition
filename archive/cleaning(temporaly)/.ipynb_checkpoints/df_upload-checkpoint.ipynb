```python
def plot_constellation_centered(frame, markdown, center_ra, center_dec):
    """
    frame: DataFrame with 'ra', 'dec', 'bf' columns, indexed by HIP
    markdown: list of tuples (hip1, hip2)
    center_ra: float, Right Ascension in hours
    center_dec: float, Declination in degrees
    """

    def sph_to_cart(ra_deg, dec_deg):
        ra_rad = np.radians(ra_deg)
        dec_rad = np.radians(dec_deg)
        x = np.cos(dec_rad) * np.cos(ra_rad)
        y = np.cos(dec_rad) * np.sin(ra_rad)
        z = np.sin(dec_rad)
        return np.stack([x, y, z], axis=0)

    def get_rotation_matrix(center_ra_deg, center_dec_deg):
        ra0 = np.radians(center_ra_deg)
        dec0 = np.radians(center_dec_deg)

        cz = np.sin(dec0)
        cx = np.cos(dec0) * np.cos(ra0)
        cy = np.cos(dec0) * np.sin(ra0)
        z_axis = np.array([cx, cy, cz])

        up = np.array([0, 0, 1])
        x_axis = np.cross(up, z_axis)
        x_axis /= np.linalg.norm(x_axis)
        y_axis = np.cross(z_axis, x_axis)

        return np.vstack([x_axis, y_axis, z_axis])  # shape: (3,3)

    center_ra_deg = center_ra * 15.0
    rotation = get_rotation_matrix(center_ra_deg, center_dec)

    plt.figure(figsize=(7, 7), dpi=100)
    ax = plt.gca()
    ax.set_facecolor('black')
    plt.axis('equal')

    for hip1, hip2 in markdown:
        stars = frame.loc[[hip1, hip2]]
        ra_deg = stars['ra'].values * 15
        dec_deg = stars['dec'].values
        labels = stars['bf'].values

        vectors = sph_to_cart(ra_deg, dec_deg)
        rotated = rotation @ vectors

        if np.all(rotated[2] < 0):
            continue

        x, y = rotated[0], rotated[1]
        plt.plot(x, y, color='white', linewidth=1)
        plt.scatter(x, y, color='white', s=10, edgecolor='white')

        # for xi, yi, label in zip(x, y, labels):
        #     plt.text(xi, yi, label, fontsize=8, color='cyan')

    plt.gca().invert_xaxis()
```
